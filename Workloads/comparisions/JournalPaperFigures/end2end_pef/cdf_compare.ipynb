```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Read the CSV files
k8s_Burstable = pd.read_csv('k8s_composePost.csv')
NetMARKS = pd.read_csv('NetMARKS_composePost.csv')
TraDE = pd.read_csv('TraDE_composePost.csv')


# Extract the 'duration' data
k8s_durations = k8s_Burstable['duration']
NetMARKS_durations = NetMARKS['duration']
TraDE_durations = TraDE['duration']

# Function to compute the CDF
def compute_cdf(data):
    sorted_data = np.sort(data)
    cdf = np.arange(1, len(sorted_data)+1) / len(sorted_data)
    return sorted_data, cdf

# Compute the CDF for each dataset
k8s_sorted, k8s_cdf = compute_cdf(k8s_durations)
NetMARKS_sorted, NetMARKS_cdf = compute_cdf(NetMARKS_durations)
TraDE_sorted, TraDE_cdf = compute_cdf(TraDE_durations)

# Plot the CDFs
plt.figure(figsize=(8, 6))
plt.plot(k8s_sorted, k8s_cdf, label='K8s Burstable')
plt.plot(NetMARKS_sorted, NetMARKS_cdf, label='NetMARKS')
plt.plot(TraDE_sorted, TraDE_cdf, label='TraDE')
plt.xlabel('Durations (ms)', fontsize =18)
plt.xticks(fontsize =15)
plt.yticks(fontsize =15)

plt.xlim(0,800)
plt.ylabel('Cumulative Distribution Function', fontsize =16)
# plt.title('CDF of Durations for K8s Burstable, NetMARKS, and TraDE')
plt.legend(loc='center right',fontsize =15)
plt.grid(True)
plt.show()
```
