```python
import matplotlib.pyplot as plt
import numpy as np

# Data of compose-post workload
policy = ['K8s-burstable', 'NetMARKS', 'TraDE']
QPS1 = [50, 200, 400]
throughput_1 = {QPS1[0]: [47, 48, 50], QPS1[1]: [92, 100, 150],  QPS1[2]: [79, 110, 155]}

# Extract throughput for each QPS
throughput_1_QPS1 = throughput_1[QPS1[0]]
throughput_1_QPS2 = throughput_1[QPS1[1]]
throughput_1_QPS3 = throughput_1[QPS1[2]]

# Combine throughput into a list for easier plotting
throughputs = [throughput_1_QPS1, throughput_1_QPS2, throughput_1_QPS3]

# Plotting
fig, ax = plt.subplots(figsize=(8, 4))

bar_width = 0.2
x = np.arange(len(QPS1))  # The label locations
# hatch_patterns = ['/', 'o', 'x']  # Hatch styles for bars
hatch_patterns = ['||', '||', '++']  # Hatch styles for bars

# Plot bars with hatching and color
for i, policy_name in enumerate(policy):
    ax.bar(x + i * bar_width, [tp[i] for tp in throughputs], width=bar_width, label=policy_name,
           color=['blue', 'orange', 'green'][i], hatch=hatch_patterns[i])

# Labeling
ax.set_xlabel('Varying QPS', fontsize=18)
ax.set_ylabel('Average Throughput (req/s)', fontsize=15)
ax.set_xticks(x + bar_width)
ax.set_xticklabels(QPS1, fontsize=15)

# Adjust tick labels
plt.yticks(fontsize=15)

# Legend
ax.legend(fontsize=14)

# Adjust layout for better fit
plt.tight_layout()

# Display the plot
plt.show()
```
