```python
import matplotlib.pyplot as plt

# Data for all three methods
categories = ['OK', '3xx', '4xx', '5xx', 'NR']
values_k8s = [65.43, 0, 0, 34.28, 0.29]
values_netmarks = [71.99, 0, 0, 27.64, 0.37]
values_trade = [95.36, 0, 0, 4.27, 0.38]

# Colors for all categories
colors = ['#228B22', '#1E90FF', '#FF4500', '#8B0000', '#000000']

# Create figure and subplots
fig, axes = plt.subplots(3, 1, figsize=(8, 4), sharex=True)  # Share x-axis for easier comparison

# Plot for K8S
axes[0].barh(categories, values_k8s, color=colors)
axes[0].set_xlim(0, 100)
axes[0].set_xticks([0, 25, 50, 75, 100])
axes[0].set_xticklabels(['0%', '25%', '50%', '75%', '100%'], fontsize=12)
axes[0].set_yticks(range(len(categories)))
axes[0].set_yticklabels(categories, fontsize=10)
# axes[0].set_ylabel("HTTP Response (K8S)", fontsize=12)
for bar, value in zip(axes[0].patches, values_k8s):
    axes[0].text(value + 1, bar.get_y() + bar.get_height() / 2, f'{value}%', va='center', fontsize=10)
axes[0].grid(axis='x', color='gray', linestyle='--', linewidth=0.5)
axes[0].set_title('(a) HTTP Response Percentages of K8s-burstable', fontsize=12)

# Plot for NetMARKS
axes[1].barh(categories, values_netmarks, color=colors)
axes[1].set_xlim(0, 100)
axes[1].set_yticks(range(len(categories)))
axes[1].set_yticklabels(categories, fontsize=10)
# axes[1].set_ylabel("HTTP Response (NetMARKS)", fontsize=12)
axes[1].set_ylabel("HTTP Response Codes", fontsize=12)

for bar, value in zip(axes[1].patches, values_netmarks):
    axes[1].text(value + 1, bar.get_y() + bar.get_height() / 2, f'{value}%', va='center', fontsize=10)
axes[1].grid(axis='x', color='gray', linestyle='--', linewidth=0.5)
axes[1].set_title('(b) HTTP Response Percentages of NetMARKS', fontsize=12)

# Plot for TraDE
axes[2].barh(categories, values_trade, color=colors)
axes[2].set_xlim(0, 100)
axes[2].set_yticks(range(len(categories)))
axes[2].set_yticklabels(categories, fontsize=10)
# axes[2].set_ylabel("HTTP Response (TraDE)", fontsize=12)
for bar, value in zip(axes[2].patches, values_trade):
    axes[2].text(value + 1, bar.get_y() + bar.get_height() / 2, f'{value}%', va='center', fontsize=10)
axes[2].grid(axis='x', color='gray', linestyle='--', linewidth=0.5)
axes[2].set_title('(c) HTTP Response Percentages of TraDE', fontsize=12)

# Shared x-axis label
# fig.text(0.5, 0.04, 'HTTP Response Percentages', ha='center', fontsize=14)

# Adjust layout
plt.tight_layout(rect=[0, 0.05, 1, 1])

# Display the plot
plt.show()
```
