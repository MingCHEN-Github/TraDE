```python
'''Plot1 (compose-post): draw comparision for average response time under differernt policy'''
import matplotlib.pyplot as plt
import numpy as np

# Data
policy = ['k8s-burstable', 'NetMARKS', 'TraDE']
QPS1 = [50, 200, 400]
response_time_1 = {QPS1[0]: [510, 390, 385], QPS1[1]: [1010, 810, 710],  QPS1[2]: [1400, 800, 700]}
QoS_target = 800


# Extract response times for each QPS
response_time_QPS1 = response_time_1[QPS1[0]]
response_time_QPS2 = response_time_1[QPS1[1]]
response_time_QPS3 = response_time_1[QPS1[2]]

# Combine response times into a list for easier plotting
response_times = [response_time_QPS1, response_time_QPS2, response_time_QPS3]

# Plotting
fig, ax = plt.subplots()

bar_width = 0.2
x = np.arange(len(QPS1))  # The label locations

# Plot bars
for i, policy_name in enumerate(policy):
    ax.bar(x + i * bar_width, [rt[i] for rt in response_times], width=bar_width, label=policy_name)
    
# Add a red dashed line for the response time target
ax.axhline(y=QoS_target, color='red', linestyle='--', linewidth=2, label=f'QoS Target ({QoS_target} ms)')

# Labeling
ax.set_xlabel('Varing QPS', fontsize=18)
ax.set_ylabel('Average Response Time (ms)',fontsize=15)

# ax.set_title('Average Response Time under Different Policies')
ax.set_xticks(x + bar_width)
ax.set_xticklabels(QPS1, fontsize = 15)
ax.set_ylim(0, 1800)
ax.legend(fontsize=14)

# Display the plot
plt.yticks(fontsize=15)

plt.tight_layout()
plt.show()
```
