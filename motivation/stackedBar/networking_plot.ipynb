```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the data directly with newlines in the application names
data = {
    'Applications': [
        'NGINX\n(monolithic)', 
        'Memcached\n(monolithic)', 
        'MongoDB\n(monolithic)', 
        'Social Network\n(Microservice)'
    ],
    'Application processing': [94.70, 80.20, 86.40, 63.70],
    'Network processing': [5.30, 19.80, 13.60, 36.30]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Normalize the percentages
df['Application processing'] = df['Application processing'] / 100.0
df['Network processing'] = df['Network processing'] / 100.0

# Plotting the stacked bar chart with vertical bars
fig, ax = plt.subplots(figsize=(10, 6))
bars = df.set_index('Applications').plot(kind='bar', stacked=True, color=['#1f77b4', '#ff7f0e'], edgecolor='black', ax=ax)

# Apply different hatch style for Social Network
bars.patches[3].set_hatch('//')  # App processing for Social Network
bars.patches[7].set_hatch('//')  # Network processing for Social Network

# Rotating the x-axis labels by 0 degrees
plt.xticks(rotation=0, ha='center', fontsize=20)
plt.yticks(fontsize=18)

# Adding y-tick labels with percentages
ax.set_yticklabels([f'{int(y * 100)}%' for y in ax.get_yticks()])

# Explicitly setting the x-axis label to an empty string
ax.set_xlabel('')

plt.ylabel('Time Percentage', fontsize=20)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=2, fontsize=18)
plt.tight_layout()  # Adjust layout to make room for rotated x-axis labels
plt.show()
```
