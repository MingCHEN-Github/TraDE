```python
import matplotlib.pyplot as plt
import numpy as np

# Please get these data from Istio service mesh, which can be viewed through Kiali dashboard
# Data 
requests = ["Compose-Post", "Read-homeTimeline", "Read-userTimeline"]
p99_latency_with_mesh = [3.74, 2.3, 2.23]
p99_latency_without_mesh = [3.6, 2.12, 2.21]


# Convert data to match the example's scale
p99_with_mesh_scaled = [latency  for latency in p99_latency_with_mesh]
p99_without_mesh_scaled = [latency  for latency in p99_latency_without_mesh]

# Configuration for bar width and positions
bar_width = 0.35  # Width of the bars
index = np.arange(len(requests))  # Numeric array for bar positioning

plt.figure(figsize=(12, 6))
bars1 = plt.bar(index, p99_with_mesh_scaled, bar_width, label='With Service Mesh', color='blue')
bars2 = plt.bar(index + bar_width, p99_without_mesh_scaled, bar_width, label='Without Service Mesh', color='green')

# plt.xlabel('Request Type (with QPS = 300)', fontsize="25")
plt.ylabel('P99 Latency (*10^2 ms)', fontsize="22")
plt.xticks(index + bar_width / 2, requests, fontsize="19", rotation=5)
plt.yticks(fontsize="18")
plt.grid(visible=True, which="major", axis="y")
plt.legend(fontsize="20")
plt.ylim([0, 5])

plt.show()

 
```
